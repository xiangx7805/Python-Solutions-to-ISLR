* the formula will look like :

$$p(Y  = \text{recieve an A}) =logit ^{-1} \left(   \hat\beta_0 + \hat\beta_1 \cdot X_1 + \hat\beta_2 \cdot X_2 \right)$$
$$ = logit ^{-1} \left(-6 + 0.05 \cdot X_{\text {Hours studied}} + 1\cdot X_{\text {undergrad GPA}}\right)$$
$$= \frac{e^{-6 + 0.05 \cdot X_{\text {Hours studied}} + 1\cdot X_{\text {undergrad GPA}}}}{1 + e^{-6 + 0.05 \cdot X_{\text {Hours studied}} + 1\cdot X_{\text {undergrad GPA}}}}$$

(a) so when $\text {Hours studied} = 40$, $\text {undergrad GPA} = 3.5$, 

we have $$p(Y  = \text{recieve an A}) = \frac{e^{-6 + 0.05 \times 40 + 1 \times 3.5}}{1 + e^{-6 + 0.05 \times 40 + 1 \times 3.5}} = 0.3775$$

(b) we want to make $p(Y  = \text{recieve an A}) > 0.5$ 

$$\Rightarrow  \frac{e^{-6 + 0.05 \times X_1 + 1 \times X_2}}{1 + e^{-6 + 0.05 \times X_1 + 1 \times X_2}} = \frac{1}{2}$$,

$$\Leftrightarrow e^{-6 + 0.05 \times X_1 + 1 \times X_2} =1$$

$$\Leftrightarrow -6 + 0.05 \times X_1 + 1 \times X_2 = 0$$

and we already have $X_2 = 3.5$, so $X_1 = 50$.

To have a 50% chance of getting an A in the class a student needs to study **50 hours**.

# Q7
![](t7.PNG)

Now we have :
$$Pr(X \mid Y = Yes)  =N(\mu = 10, \hat \sigma^2 = 36)  = f_{Yes}(x)$$
$$Pr(X \mid Y = No) = N(\mu = 0, \hat \sigma^2 = 36) = f_{No}(x)$$
$$Pr(Y = Yes) = \pi_{Yes} = 0.8 $$
$$Pr(Y = No) = \pi_{No} = 0.2 $$

We want to calculate $$Pr(Y =Yes \mid X=4)$$ using *Bayes' theorem*.

$$Pr(Y =Yes \mid X=4) = \frac{Pr(Y =Yes , X=4)}{Pr(X=4)}$$
$$= \frac{Pr(Y =Yes , X=4)}{Pr(Y =Yes , X=4) + Pr(Y =No , X=4)}$$

$$= \frac{Pr(Y =Yes , X=4)}{Pr( X=4 \mid Y =Yes ) Pr(Y = Yes) + Pr(X=4 \mid Y =No) Pr(Y = No)}$$
$$= \frac{Pr(Y =Yes , X=4)}{Pr( X=4 \mid Y =Yes ) Pr(Y = Yes) + Pr(X=4 \mid Y =No) Pr(Y = No)}$$
$$= \frac {Pr( X=4 \mid Y =Yes ) Pr(Y = Yes)}{Pr( X=4 \mid Y =Yes ) Pr(Y = Yes) + Pr(X=4 \mid Y =No) Pr(Y = No)}$$

$$= \frac{0.8\times e^{-(4-10)^2/(2*36)} }{0.8\times e^{-(4-10)^2/(2*36)} + 0.2\times e^{-(4-0)^2/(2*36)}} = 0.7571$$

## (c) confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
weekly.head()
```

```python
weekly = pd.read_csv('Weekly.csv',index_col = 0)
pred_col = ['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']
lr = LogisticRegression()
reg = lr.fit(X = weekly[pred_col],y = weekly["Direction"])

# confusion matrix
conf_mat = confusion_matrix(weekly['Direction'],lr.predict(weekly[pred_col]))
print(conf_mat)
```

```python
#'fancy' confusion matrix plot
#based on: Raschka (2014)
fig, ax = plt.subplots(figsize=(2, 2))
ax.matshow(conf_mat, cmap=plt.cm.Reds, alpha=0.3)
for i in range(conf_mat.shape[0]):
    for j in range(conf_mat.shape[1]):
        ax.text(x=j, y=i,
        s=conf_mat[i, j],
        va='center', ha='center')
plt.xlabel('predicted label')
plt.ylabel('true label')
plt.show()
```

The confusion matrix gives us two types of mistakes: 
    1) false positives; 
    2) false negatives.
    
> A **false positive** occurs when our prediction gives us a positive value but the real value is negative. 

> A **false negative** occurs when our prediction gives us a negative value but the real value is positive.

Assuming that class 1 ('Up') is the positive class, our model correctly classified 55 samples that belong to class 0 (true negatives) and 558 samples that belong to class 1 (true positives). However, our model also incorrectly misclassified 429 samples from class 0 to class 1 (false negatives), and it predicted that 47 samples were 'Up' although they were 'Down' (false positives).

```python
#overall fraction of correct predictions
print('overall fraction of correct predictions is about ',round(lr.score(weekly[pred_col],weekly['Direction']),3))
```

```python
# (e)
def PLOT(log = ''):
    x= np.arange(1,10)
    y = power3(x,2)
    
    fig,ax = plt.subplots()
    
    # config plot
    ax.set_xlabel('x')
    ax.set_ylabel('y=x^2')
    ax.set_title('power3( )')
    
    #change scale to axis
    if log == 'x':
        ax.set_xscale('log')
        ax.set_xlabel('log(x)')
        
    if log =='y':
        ax.set_yscale('log')
        ax.set_ylabel('log(y = x^2)')
        
    if log == 'xy':
        ax.set_xscale('log')
        ax.set_xlabel('log(x)')
        ax.set_yscale('log')
        ax.set_ylabel('log(y = x^2)')
    
    #draw
    ax.plot(x,y)
```

```python
def PlotPower(start, end, power, log = ''):
    x = np.arange(start, end)
    y = np.power(x,power)
    
    fig,ax = plt.subplots()
    
    ax.set_xlabel('x')
    ax.set_ylabel('y=x^{}'.format(power))
    ax.set_title('PlotPower( )')
    
    if log == 'x':
        ax.set_xscale('log')
        ax.set_xlabel('log(x)')
        
    if log =='y':
        ax.set_yscale('log')
        ax.set_ylabel('y=x^{}'.format(power))
        
    if log == 'xy':
        ax.set_xscale('log')
        ax.set_xlabel('log(x)')
        ax.set_yscale('log')
        ax.set_ylabel('y=x^{}'.format(power))
    
    #draw
    ax.plot(x,y)
```

```python
boston = pd.read_csv('Boston.csv', index_col = 0)
boston.head()
```

```python
boston.describe()
```

```python
# boston = load_boston()
# df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
boston.columns.values
```

```python
# create new variable
boston['crim01'] = np.where(boston['crim'] > boston['crim'].median(),1,0)
boston.head()
```

```python
boston.drop('crim', inplace=True, axis=1)
boston.head()
```

```python
boston.corr()
```

```python
list(boston.columns.values)
```

```python
x_all = boston.values
x_sub = boston[['indus','nox', 'age', 'dis', 'rad', 'tax', 'medv']].values
y = boston['crim01'].values

x_all_train, x_all_test, y_train, y_test = train_test_split(x_all, y, random_state=1)

x_sub_train, x_sub_test, y_train, y_test = train_test_split(x_sub, y, random_state=1)
```

## Logistic Regression

```python
lr = LogisticRegression()
lr.fit(x_sub_train, y_train)
print("Accuracy with x_sub  :",round(accuracy_score(y_test, lr.predict(x_sub_test)),3))
lr.fit(x_all_train, y_train)
print("Accuracy with x_all:",round(accuracy_score(y_test, lr.predict(x_all_test)),3))
```

## LDA

```python
lda = LinearDiscriminantAnalysis()
lda.fit(x_sub_train, y_train)
print("Accuracy with x_sub  :",round(accuracy_score(y_test, lda.predict(x_sub_test)),3))
lda.fit(x_all_train, y_train)
print("Accuracy with x_all:",round(accuracy_score(y_test, lda.predict(x_all_test)),3))
```

## QDA

```python
qda = QuadraticDiscriminantAnalysis()
qda.fit(x_sub_train, y_train)
print("Accuracy with x_sub  :",round(accuracy_score(y_test, qda.predict(x_sub_test)),3))
qda.fit(x_all_train, y_train)
print("Accuracy with x_all:",round(accuracy_score(y_test, qda.predict(x_all_test)),3))
```

## KNN with best K

```python
best_acc = 0
best_k = 0
for K in range(1,101):
    knn = KNeighborsClassifier(n_neighbors=K)
    knn.fit(x_all_train, y_train)
    acc = accuracy_score(y_test, knn.predict(x_all_test))
    if acc > best_acc:
        best_acc, best_k = acc, K

print('Best accuracy = {:.4f} with K = {:3}'.format(best_acc, best_k))
```
